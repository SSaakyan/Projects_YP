## Описание данных

Датасет описывает транзакции интернет-магазина.

Колонки в  *ecommerce_dataset.csv* :

- **date** — дата заказа;
- **customer_id** — идентификатор покупателя;
- **order_id** — идентификатор заказа;
- **product** — наименование товара;
- **quantity** — количество товара в заказе;
- **price** — цена товара.

- Для дальнейшего исследования, сосчитаем выручку по каждой позиции:

```python
df['revenue'] = df['quantity'] * df['price']
```

**1. Проверим, есть ли в данных заказы с одним номером, но в разные дни.**

```python
df.groupby(['order_id', 'date'])['revenue'].sum()
```

Видим, что есть заказы с одним и тем же номером, одной и той же суммой, но с разным временем покупки. Вероятно, это одни и те же заказы с разным статусом (например, 'сформирован', 'в статусе оформления', 'оплата' и т.п.) или это тестовые заказы или техн.ошибки (стоит передать информацию коллегам). Проверим сколько таких в нашем датасете.

```python
df1 = df.groupby('order_id').agg({'date' : 'nunique', 'customer_id' : 'count'}).query('date > 1')
df1
```

Количество позиций в одном и том же заказе менялось (например, заказ 14500). Рассчитаем приблизительно долю строк в датасете, которые были в дублирующихся заказах.

Добавим дополнительный столбец, в котором будет указано сколько позиций было в одном заказе (расчёт приблизительный, надо понимать порядок процента таких строк в нашем датасете)

```python
df1['products_by_order'] = df1['customer_id'] / df1['date']
```

Рассчитаем процент строк с дублирующими заказами без учёта каждого заказа один раз (т.е. если бы мы хотели удалить строки с дублирующими заказами, то оставили бы только заказы, сделанные последними по времени)

```python
print('Процент строк с дублирующимися заказами: {:.2%}'
      .format((df1['customer_id'].sum() - df1['products_by_order'].sum()) / len(df))
     )
```

Итак, у нас есть  284 заказа с уникальным номером, но в данных они присутствуют в разные даты. У нас нет дополнительной информации, были ли это тестовые заказы или это одни и те же заказы с разными статусами, но процент дублирующихся заказов большой, поэтому удалять их не будем, хоть это и даёт некоторый процент искажения.

**2. Проверим, есть ли в данных заказы с одним номером,но с разными идентификаторами покупателей.**

```python
df2 = df.groupby('order_id').agg({'customer_id' : 'nunique', 'product' : 'count'}).query('customer_id > 1')
df2.shape
```

```python
df2
```

Сосчитаем процент таких дублирующих строк в датасете (без учёта сделанного последним заказа). Для этого также введём новый столбец, в  котором будет указано сколько позиций было на одного покупателя (расчёт приблизительный, надо понимать порядок процента таких строк в нашем датасете)

```python
df2['products_by_customer'] = df2['product'] / df2['customer_id']
```

```python
print('Процент строк с дублирующимися заказами: {:.2%}'
      .format((df2['product'].sum() - df2['products_by_customer'].sum()) / len(df))
     )
```

Процент маленький, поэтому удалим дублирующие строки, кроме заказов, сделанных последними (для тех заказов, у которых покупатель не один)

```python
#Добавим дополнительный столбец с последней датой заказа для каждого заказа c разными покупателями:
last_order_date = (
    df.groupby('order_id')
    .agg({'customer_id' : 'nunique', 'date': 'max'})
    .query('customer_id > 1')
    .drop('customer_id', axis=1)
)
last_order_date.columns = ['last_order']
last_order_date.name = 'last_order_date'
df = df.join(last_order_date, on='order_id', how='left')
```

Пустые значения заменим на датой из date

```python
df['last_order'] = df['last_order'].fillna(df['date'])
```

Теперь удалим дублирующие строки. Затем удалим вспомогательный столбец last_order

```python
df = df.query('date == last_order').drop('last_order', axis=1)
```

### Анализ выручки, количества покупателей и среднего чека в динамике

#### В зависимости от дня заказа

```python
#рассчитываем выручку и количество покупателей в зависимости от дня заказа:
df_grouped_by_order_day = (
    df.groupby('day')
    .agg({'revenue': 'sum', 'order_id': 'nunique'})
    .reset_index()
)

#рассчитываем средний чек в зависимости от дня заказа:
df_grouped_by_order_day['revenue_per_order'] = (
    df_grouped_by_order_day['revenue'] / df_grouped_by_order_day['order_id']
)
```

```python
#рисуем графики:
fig, ax = plt.subplots(3 ,1, figsize=(18,16))

sns.lineplot(x=df_grouped_by_order_day['day'],y=df_grouped_by_order_day['revenue'],ax=ax[0])
sns.lineplot(x=df_grouped_by_order_day['day'],y=df_grouped_by_order_day['order_id'],ax=ax[1])
sns.lineplot(x=df_grouped_by_order_day['day'],y=df_grouped_by_order_day['revenue_per_order'],ax=ax[2])

ax[0].set_title('Выручка по дням', size=16)
ax[1].set_title('Количество уникальных покупателей по дням', size=16)
ax[2].set_title('Средний чек по дням', size=16)

ax[0].tick_params(axis='x', labelrotation=30)
ax[1].tick_params(axis='x', labelrotation=30)
ax[2].tick_params(axis='x', labelrotation=30)

ax[0].set_xlabel('Дата',size=12)
ax[0].set_ylabel('Выручка',size=14)

ax[1].set_xlabel('Дата',size=12)
ax[1].set_ylabel('Кол-во уникальных покупателей',size=14)

ax[2].set_xlabel('Дата',size=12)
ax[2].set_ylabel('Средний чек',size=14)

fig.tight_layout()
```

Автоматизируем построение графиков выручки, количества покупателей и среднего чека по часу, дням недели, неделям, месяцам:

```python
def graf(time, title, rotation, xtitle, size,titlesize,labelsize,tickssize):
    
    #группируем данные по времени заказа:
    df_grouped_by_order_time = (
        df.groupby(time)
        .agg({'revenue': 'sum', 'order_id': 'nunique', 'customer_id' : 'nunique'})
        .reset_index()
        )
    
    #расчитываем средний чек:
    df_grouped_by_order_time['revenue_per_order'] = (
        df_grouped_by_order_time['revenue'] / df_grouped_by_order_time['order_id']
    )
    
    #рисуем графики:
    fig, ax = plt.subplots(3 ,1, figsize=size)
    
    sns.barplot(x=df_grouped_by_order_time[time],y=df_grouped_by_order_time['revenue'], ax=ax[0])
    sns.barplot(x=df_grouped_by_order_time[time],y=df_grouped_by_order_time['customer_id'], ax=ax[1])
    sns.barplot(x=df_grouped_by_order_time[time],y=df_grouped_by_order_time['revenue_per_order'], ax=ax[2])
    
    ax[0].set_title('Выручка ' + title, size=titlesize)
    ax[1].set_title('Количество уникальных покупателей ' + title, size=titlesize)
    ax[2].set_title('Средний чек ' + title, size=titlesize)

    ax[0].tick_params(axis='x', labelrotation=rotation)
    ax[1].tick_params(axis='x', labelrotation=rotation)
    ax[2].tick_params(axis='x', labelrotation=rotation)

    ax[0].set_xlabel(xtitle,size=labelsize)
    ax[0].set_ylabel('Выручка',size=labelsize)

    ax[1].set_xlabel(xtitle,size=labelsize)
    ax[1].set_ylabel('Кол-во уникальных покупателей',size=labelsize)

    ax[2].set_xlabel(xtitle,size=labelsize)
    ax[2].set_ylabel('Средний чек',size=labelsize)
    
    ax[0].tick_params(axis='x', labelsize=tickssize)
    ax[1].tick_params(axis='x', labelsize=tickssize)
    ax[2].tick_params(axis='x', labelsize=tickssize)
    
    ax[0].tick_params(axis='y', labelsize=tickssize)
    ax[1].tick_params(axis='y', labelsize=tickssize)
    ax[2].tick_params(axis='y', labelsize=tickssize)
      
    fig.tight_layout()
    return plt.show()
```

## Проверка гипотез

### Гипотеза 1

**Нулевая гипотеза H₀**: *средняя выручка по товарам кластера 0 равна средней выручке по товарам кластера 1.*

**Альтернативная гипотеза H₁**: *средняя выручка по товарам кластера 0 больше средней выручки по товарам кластера 1 .*

Чтобы проверить гипотезу о средних воспользуемся t-тестом.

```python
alpha = 0.05 # критический уровень статистической значимости
```

```python
level = 388 #средняя выручка по кластеру 1
```

```python
results = st.ttest_1samp(features.query('cluster_km == 0')['revenue'], level)
```

```python
print('p-значение: ', results.pvalue / 2)
if (results.pvalue / 2 < alpha) and (features.query('cluster_km == 0')['revenue'].median() > level):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод:** Отвергаем нулевую гипотезу, принимаем альтернативную гипотезу: что средняя выручка по товарам кластера 0 больше средней выручки по товарам кластера 1.
