## Введение

### Описание данных

**Структура visits_log.csv** (хранит лог сервера с информацией о посещениях сайта)

- *Uid* — уникальный идентификатор пользователя,
- *Device* — категория устройства пользователя,
- *Start Ts* — дата и время начала сессии,
- *End Ts* — дата и время окончания сессии,
- *Source Id* — идентификатор источника перехода на сайт.

**Структура orders_log.csv** (c информацией о заказах)

- *Uid* — уникальный идентификатор пользователя,
- *Buy Ts* — дата и время заказа,
- *Revenue* — сумма заказа.

**Структура costs.csv** (с информацией о расходах на маркетинг)

- *source_id* — идентификатор рекламного источника,
- *dt* — дата проведения рекламной кампании,
- *costs* — расходы на эту кампанию.

Выручка измеряется в условных единицах — у.е.
Маржинальность сервиса 100%

## Расчёт метрик и построение графиков

### Продуктовые метрики

#### DAU, WAU, MAU

```python
# Для расчётов возьмём те посещения, где время окончания сессии поднее времени начала сессии
visits_research = visits[visits['start_ts'] < visits['end_ts']]
```

```python
import sys
import warnings
if not sys.warnoptions:
       warnings.simplefilter("ignore")
```

```python
# Для дальнейших вычислений выделим дату, неделю, месяц и год в отдельные столбцы
visits['session_date'] = visits['start_ts'].dt.date
visits['session_week'] = visits['start_ts'].astype('datetime64[W]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits['session_year'] = visits['start_ts'].dt.year
```

##### DAU

Посчитаем метрику и ее среднее значения за весь период:

```python
dau = visits.groupby('session_date').agg({'uid':'nunique'}).reset_index()
dau_total = dau['uid'].mean()
print('Среднее количество уникальных пользователей в день: ', round(dau_total,2))
```

Теперь взглянем на изменение метрики во времени:

```python
plt.figure(figsize=(18,6))
plt.title("Количество уникальных пользователей в день",size=18)
sns.lineplot(x=dau['session_date'], y=dau['uid']).grid()
plt.xlabel("День",size=14)
plt.xticks(rotation=15,size=8)
plt.ylabel("Количество уникальных пользователей",size=14)
plt.show()
```

**Вывод:** Количество новых пользователей в день растёт в осенние месяцы и постепенно снижается с февраля 2018г.Есть три самых больших выброса: 24.11.2017, 31.03.2018 и конец мая-начало июня 2018г. Рост пользователей идёт после запуска рекламных кампаний.

```python
# посмотрим, в какие даты происходят выбросы
print(
    'Максимальное количество уникальных пользователей равное {} было {}'
    .format(dau['uid'].max(), 
            dau.loc[dau['uid'] == dau['uid'].max(),'session_date']
            .to_string(index=False))
)
```

```python
print(
    'Минимальное количество уникальных пользователей равное {} было {}'
    .format(dau['uid'].min(), 
            dau.loc[dau['uid'] == dau['uid'].min(),'session_date']
            .to_string(index=False))
)
```

**Вывод:**
Количество новых пользователей в день растёт в осенние месяцы и постепенно снижается с февраля 2018г.Есть три самых больших выброса: 24.11.2017, 31.03.2018 и конец мая-начало июня 2018г. (данные есть только до конца мая 2018г.)

##### WAU

Посчитаем метрику и ее среднее значения за весь период:

```python
wau =  visits.groupby(['session_week']).agg({'uid':'nunique'}).reset_index()
wau_total = wau['uid'].mean()
print('Среднее количество уникальных пользователей в неделю: ', round(wau_total,2))
```

Посмотрим на изменение метрики во времени:

```python
plt.figure(figsize=(24,10))
plt.title("Количество уникальных пользователей в неделю",size=18)
sns.lineplot(
    y=wau['uid'],
    x=wau['session_week'].astype(str)
).grid()

plt.xlabel("Неделя",size=16)
plt.xticks(rotation=90,size=12)
plt.ylabel("Количество уникальных пользователей",size=16)
plt.yticks(size=12)
plt.show()
```

**Вывод:** наибольшее значение наблюдается 23-29 ноября 2017г. Минимальное число новых пользователей было с 3 по 9 августа 2017г. (31 мая 2018 -один день, поэтому его некорректно брать в расчёт). С 23 ноября 2017г. наблюдается тренд на снижение.

```python
mau = visits.groupby(['session_month']).agg({'uid':'nunique'}).reset_index()
mau_total = mau['uid'].mean()
print('Среднее количество уникальных пользователей в месяц: ', round(mau_total,2))
```

```python
plt.figure(figsize=(20,10))
plt.title("Количество уникальных пользователей в месяц",size=18)
sns.lineplot(
    y=mau['uid'],
    x=mau['session_month'].astype(str),
).grid()

plt.xlabel("Месяц",size=16)
plt.xticks(rotation=30,size=12)
plt.ylabel("Количество уникальных пользователей",size=16)
plt.yticks(size=12)
plt.show()
```

#### Retention Rate

Приведём тип данных в столбце session_date к datetime:

```python
visits['session_date'] = visits['session_date'].astype('datetime64')
```

Найдём даты, когда каждый пользователь впервые зашел на сайт, и добавим их в датафрейм c информацией о посещениях сайта:

```python
first_activity_date = visits.groupby(['uid'])['start_ts'].min()
first_activity_date.name = 'first_activity_date'
visits = visits.join(first_activity_date, on='uid')
```

Формировать когорты будем по данным за месяц. 

```python
visits['first_month'] = visits['first_activity_date'].astype('datetime64[M]')
```

Рассчитаем lifetime:

```python
visits['cohort_lifetime'] = (
(visits['session_month'] - visits['first_month']) / np.timedelta64(1,'M')
).round().astype('int')
display(visits.head())
```

Группируем данные по когорте и lifetime:

```python
cohorts = visits.\
groupby(['first_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
display(cohorts.head())
```

Найдём число пользователей на нулевой месяц для каждой когорты:

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][
    ['first_month', 'uid']
]
```

Для корректного объединения таблиц переименуем столбец uid в cohort_users:

```python
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
display(initial_users_count.head())
```

Объединим данные по когортам с числом пользователей на 0-ом месяце:

```python
cohorts = cohorts.merge(initial_users_count,on='first_month') 
display(cohorts.head())
```

Найдем Retention Rate:

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 
```

```python
# месяц будем выводить в формате год-месяц
cohorts['first_month'] = cohorts['first_month'].dt.strftime('%Y-%m')
```

Нарисуем тепловую карту, для этого построим сводную таблицу:

```python
retention_pivot = cohorts.pivot_table(
    index='first_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
) 
```

```python
display(retention_pivot.fillna(''))
```

Создадим тепловую карту:

```python
sns.set(style='white')
fig, ax = plt.subplots(1, 1, figsize = (24, 18))
plt.title('Retention Rate для когорт в разрезе lifetime',size=24)
sns.heatmap(
    retention_pivot,
    annot=True,
    annot_kws={"fontsize":18},
    vmin=0,
    vmax=0.1,
    fmt='.1%',
    linewidths=1,
    linecolor='grey',
    cmap= 'coolwarm',
)
plt.xticks(size=16)
plt.yticks(rotation=45,size=16)
ax.set_ylabel('Когорта', size=18)    
ax.set_xlabel('Lifetime', size=18)
plt.show()
```

#### LTV

Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку:

```python
cohorts_ltv = (
    orders_new.groupby(['first_purchase_month', 'purchase_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
display(cohorts_ltv.head()) 
```

 Добавим количество пользователей, которые первый раз совершили покупку в каждый месяц:

```python
report_ltv = pd.merge(cohort_sizes, cohorts_ltv, on='first_purchase_month')
display(report_ltv.head()) 
```

Рассчитаем валовую прибыль:

```python
margin_rate = 1
report_ltv['gp'] = report_ltv['revenue'] * margin_rate
```

Найдём lifetime:

```python
report_ltv['age'] = (
    (report_ltv['purchase_month'] - report_ltv['first_purchase_month']) / np.timedelta64(1, 'M')
).round().astype('int')
display(report_ltv.head())
```

Рассчитаем LTV:

```python
report_ltv['ltv'] = report_ltv['gp'] / report_ltv['n_buyers']
```

Для когорт будем выводить только год и месяц:

```python
report_ltv['first_purchase_month'] = report_ltv['first_purchase_month'].dt.strftime('%Y-%m')
```

Для наглядности посмотрим результаты в сводной таблице:

```python
output = report_ltv.pivot_table(
    index='first_purchase_month', columns='age', values='ltv', aggfunc='sum'
).cumsum(axis=1).round(2)
display(output.fillna(''))

```

Рассчитаем средний LTV по когортам за 6 месяцев для тех когорт, которые «прожили» не менее 6 месяцев:

```python
print(
    'Cредний LTV за 6 месяцев по когортам, «прожившим» не менее 6 месяцев равен {:.2f}'
    .format(output[5].mean())
)
```

Посмотрим на изменение метрики во времени:

```python
sns.set(style='white')
fig, ax = plt.subplots(1, 1, figsize = (20, 18))
plt.title('LTV для когорт в разрезе lifetime',size=24)
sns.heatmap(
    output,
    annot=True,
    annot_kws={"fontsize":18},
    vmin=3,
    vmax=15,
    linewidths=1,
    linecolor='grey',
    cmap= 'coolwarm',
)
plt.xticks(size=16)
plt.yticks(rotation=45,size=16)
ax.set_ylabel('Когорта', size=18)    
ax.set_xlabel('Lifetime', size=18)
plt.show()
```

**Вывод:** 
1. В среднем каждый покупатель из когорт, «проживших» не менее 6 месяцев, принес 7.97 у.е. валовой прибыли за 6 месяцев "жизни";

#### ROMI

Сосчитаем ROMI для каждого источника в цикле:

```python
#пройдёмся по всему списку уникальных источников:
for source_id in marketing['source_id'].unique():
    print('ROMI по когортам для источника', source_id)
    marketing_by_source = marketing.query('source_id == @source_id')
    orders_by_source = clients.query('source_id == @source_id')
    
    # расчёт LTV по когортам в разрезе lifetime:
    cohorts = orders_by_source.groupby(['first_purchase_month','purchase_month']).agg({'revenue': 'sum'}).reset_index()
    first_orders = orders_by_source.loc[:,['uid','first_purchase_month']]
    cohort_sizes = first_orders.groupby('first_purchase_month').agg({'uid': 'nunique'}).reset_index()
    cohort_sizes.columns = ['first_purchase_month', 'n_buyers']
    
    report = pd.merge(cohorts, cohort_sizes, on='first_purchase_month')
    report['age'] = (report['purchase_month'] - report['first_purchase_month']) / np.timedelta64(1, 'M') 
    report['age'] = report['age'].round().astype('int')
    report['ltv'] = (report['revenue'] * margin_rate) / report['n_buyers'] 
    
    # расчёт САС:
    monthly_costs = marketing_by_source.groupby('costs_month').agg({'costs': 'sum'}).reset_index()
    report_new = pd.merge(report, monthly_costs, left_on='first_purchase_month', right_on='costs_month')
    report_new['cac']  = report_new['costs'] / report_new['n_buyers']
    
    # расчёт ROMI:
    report_new['romi'] = report_new['ltv'] / report_new['cac']
    report_new['first_purchase_month'] = report_new['first_purchase_month'].dt.strftime('%Y-%m')
    romi  = report_new.pivot_table(
                    index='first_purchase_month', 
                    columns='age', 
                    values='romi', 
                    aggfunc='mean'
    ).cumsum(axis=1).round(2)
    display(romi.fillna(''))
    
    # рисуем тепловую карту:
    sns.set(style='white')
    fig, ax = plt.subplots(1, 1, figsize = (16, 14))
    plt.title('Romi для когорт в разрезе lifetime',size=14)
    
    sns.heatmap(
        romi,
        annot=True,
        annot_kws={"fontsize":12},
        vmin=0,
        vmax=1.1,
        fmt='.2f',
        linewidths=1,
        linecolor='grey',
    )
    plt.xticks(size=10)
    plt.yticks(rotation=45,size=10)
    ax.set_ylabel('Когорта', size=12)    
    ax.set_xlabel('Lifetime', size=12)
    plt.show()
    
```

2. Лучше всего обстоят дела с 1 источником: на шестой месяц "жизни" всего когорты, живущие более 6 месяцев окупились. Более того для большинства когорт у этого источника окупаемость уже на второй месяц "жизни"
